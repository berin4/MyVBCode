##datetime module
- provides full date and time
- set your timezone with **pytz** module
- "Timely Greeting" : say Good Morn/Afternoon/Eve

```python
# import the datetime package / library / object / class
import datetime
# import pytz in order to set time to your local time zone
import pytz
my_tz = pytz.timezone("America/New_York")

# instantiate datetime object called dt
# and set it for the current timezone
dt = datetime.datetime.now(my_tz) # 2024-03-19 10:19:27.874826-04:00

print(dt)
```

```python
# show all timezone codes:
# pytz.all_timezones is example of obj.property
print(len(pytz.all_timezones)) # 596
# print every 50th tz:
pytz.all_timezones[0:575:25] # [start_index:end_index_exclusive:step]
```

- **get individual time and date units**
- **dt.strftime('%char_code')** returns string version of date

```python
print('dt:', dt, type(dt)) # <class 'datetime.datetime'>

# parse the datetime object:

yr = dt.year # get the year property
print('year:', yr, type(yr)) # year: 2024 <class 'int'>

mo_num = dt.month # get the month as int
print('month as number:', mo_num)

mo_str = dt.strftime('%B') # get the name of the full month
print('full month string:', mo_str) # March

mo = dt.strftime('%b') # get the abbrev name of month
print('abbrev month string:', mo) # Mar

# get the date from 1-31
date = dt.day
print('date:', date) # 19

# get the full name day of the week
day_str = dt.strftime('%A')
print('day_str:', day_str)

# get the abbrev name day of the week
day = dt.strftime('%a')
print('day_str:', day)

# get the hour as int from 0-23
hr = dt.hour
print('hour:', hr)

# get the minute from 0-59
min = dt.minute
print('min:', min)

# get the second as int from 0-59
sec = dt.second
print('sec:', sec)

# get the full time in military format
# add leading 0 as needed
if min < 10:
  min = f'0{sec}'

if sec < 10:
  sec = f'0{sec}'

time_now = f"{hr}:{min}:{sec}"
print('time_now:', time_now)

formatted_time = dt.strftime('%H:%M:%S')
print('formatted_time:', formatted_time) # '14:30:15'

# get the full time in AM-PM format without seconds
am_pm_time = dt.strftime('%I:%M %p')
print('am_pm_time:', am_pm_time) # '2:30:15 PM'

# get today's date:
# '%B %d, %Y'
print()
```
